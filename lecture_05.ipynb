**Document Loaders in LangChain**

**Writing Document In Standard Form**

documents come in many different shapes — a PDF, a Word file, a webpage, a CSV, etc. But your RAG pipeline doesn't understand all these formats directly. So you need to bring everything into one common, simple format that the system can work with.

Raw File (PDF, URL, DOCX...)
        ↓
   Document Loader
        ↓
  Standard Document Object  ← "Standard Form"
  { page_content, metadata }
        ↓
   Split → Embed → Store in Vector DB

```python
from langchain_core.documents import Document
sample_document = Document(
    page_content = "This is the sample page content.",
    metadata ={"source":"Yar",
               "page":0,}
)

print(f"Content Of the File:{sample_document.page_content}")
print(f"metadata of the file:{sample_document.metadata}")
print(f"type of document:{type(sample_document)}")
```

```python
#To individually Access the each metadata
source = sample_document.metadata.get("source","unknown")
print(f"{source}")
```

**Txt DOcument Loader**

Simple documnet loader or raw form document loader,This loader consider the all pages as one document or 1 page cannot differentiate or recongize pages.

```python
from langchain_community.document_loaders import TextLoader
txt_file = TextLoader(
    file_path = "lang_practice.txt",
    encoding ="utf-8"
)

load_txt = txt_file.load()

print(f"Length of file is:{len(load_txt)}")
print(f"Page content is:{load_txt[0].page_content}")
print(f"metadata is:{load_txt[0].metadata}")
```
