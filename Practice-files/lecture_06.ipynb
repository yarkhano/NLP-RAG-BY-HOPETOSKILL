**Text Splitter or Chunkers**

```python
from langchain_community.document_loaders import TextLoader
txt_loader = TextLoader(
    file_path = "yar.txt"
)

txt_file = txt_loader.load()

content = txt_file[0].page_content

print(f"length of content: {len(content)}")

#Tokens that this file have
print(f"The estimated tokens are:{len(content)//4}")

#first 20 characters of the file are
print(f"first 300 characters of the file are :\n\n{content[:300]}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 200
)


chunks = splitter.split_documents(txt_file)

print(f"length of the document:{len(txt_file)}")
print(f"chunk size is:{len(chunks)}")
```

```python
for index,chunk in enumerate(chunks[:3]):
    print(f"length of chunk {index+1} :{len(chunk.page_content)}")
```

```python
# Now we have to verify the Overlapping
if len(chunks) >= 2:

    check_overlap = False

    for i in range(len(chunks) - 1):
        current_chunk = chunks[i].page_content
        next_chunk = chunks[i + 1].page_content

        if current_chunk[-200:] == next_chunk[:200]:
            check_overlap = True
            print(f"Chunks {i} and {i+1} are overlapped")
        else:
            print(f"Chunks {i} and {i+1} are NOT overlapped")
```
